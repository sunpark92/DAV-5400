```python
# import numpy and pandas libraries
import numpy as np
import pandas as pd
```

```python
# import diamonds.csv file
data=pd.read_csv('https://raw.githubusercontent.com/sunpark92/DAV-5400/master/diamonds.csv')
# print the data
data
```

### 1. What proportion of diamonds are between .30 and 1.08 carats?

```python
# name 'carat' for diamonds are between .30 and 1.08 carat
# which carat are between 0.3 and 1.08 carats and data.shape[0] to get the total number
carat=((data[(data['carat']>=0.30) & (data['carat']<=1.08)].shape[0])/(data.shape[0]))*100
carat
```

### 2. How many of the diamonds have equal x and y dimensions?

```python
# count diamonds with equal x and y dimensions
equal=data.loc[data['x'] == data['y']]
equal.x.count()
```

### 3. How many of the diamonds have a carat less than the mean carat value?

```python
# count diamonds have a carat less than the mean carat value
carat_less=data[data['carat'] < data['carat'].mean()]
carat_less.x.count()
```

### 4. How many diamonds have a Premium cut or better? Note that the possible values for the quality of a cut are ranked in ascending order as follows: Fair / Good / Very Good / Premium / Ideal

```python
# count the number of ideal cut diamond
Ideal= data[(data['cut']=='Ideal')]
Ideal.count()
```

### 5. Which diamond has the highest price per carat? What is its value?

```python
# new variable highprice
highprice = data
highprice
# create a new column call pp_carat
highprice['pp_carat'] = (highprice['price']/highprice['carat'])
value = highprice[highprice['pp_carat'] == highprice['pp_carat'].max()]
# price per carat
print('Price per carat is: ',value['pp_carat'])
# print the price
print('Price is: ',value['price'])
# print the row 
value
```
