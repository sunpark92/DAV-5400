#### Create a Jupyter Notebook that generates a text file(“tournamentinfo.txt”) as a .CSV file with the following information for all of the chess players: 

Player’s Name, Player’s State, Total Number of Points, Player’s Pre-Rating, and Average Pre Tournament Chess Rating of Opponents

__To convert data from text to csv with all information above, I have to cleanse data with the following processes.__
I am going to extract each column and then combine all columns by using dataframe. 

```python
# import csv,numpy,pandas,re library
import csv
import numpy as np
import pandas as pd
import re
```

```python
# create a empty list a to store data
wholelist = []
with open('tournamentinfo.txt') as csvfile: 
    readCSV = csv.reader(csvfile)
    for row in readCSV:
        wholelist.append(row)
# print wholelist
wholelist
```

#### 1) First, I am going to extract Player’s Name

```python
# remove first three rows
wholelist = wholelist[3:]
# use len function to get the length of wholelist
length = len(wholelist)
```

```python
# use np.arange function
# np.arange([start, ] stop, [step, ])
list1 = np.arange(1,length,3) 
aaa = 0
# create a empty list to store first arange list
# and create a for loop
first = []
for i in wholelist:
    if(aaa in list1):
        first.append(i)
    aaa = aaa+1
#print first 
first
```

```python
# create wholelist regular expression pattern and store in pn
# player's names are two words or three words
pn = re.compile('[a-zA-Z]+\s[a-zA-Z]+\s[a-zA-Z]*')
# Player's Name = PN
PN = pn.findall(str(first))
PN
```

#### 2) Second, I am going to extract Player’s State

```python
# use len function to get the length of wholelist
len_list = len(wholelist)
# use np.arange to arange wholelist
npwholelist = np.arange(2,len_list,3)
bbb = 0
# create a empty list to store first arange list
ST = []
#create a for loop
for i in wholelist:
    if(bbb in npwholelist):
        ST.append(i)
    bbb = bbb+1
#print ST
ST
```

```python
# create a regular expression pattern and store it in SN(state name)
SN = re.compile('[a-zA-Z]{2}') # repeat alphabet only 2 times
# use findall function to find all values which match the pattern
state = SN.findall(str(ST))
# print state
state
```

#### 3) Get Total Number of Points

```python
# create a regular expression pattern and store in RS(R+Score), 
RS = re.compile('[A-Z]{1}\:\s+[0-9]+')
# use findall function to find all values which match the pattern
points = RS.findall(str(ST))
# create a new regular expression pattern and store it in PO,
PO = re.compile('[0-9]+') #repeat number 
# use findall function to find all values which match the pattern
NP = PO.findall(str(points))       
# print points
NP
```

#### 4) Get Player’s Pre-Rating

```python
# create a regular expression pattern and store in pr
pr = re.compile('[0-9].[0-9]')
# use findall function to find all values which match the pattern
PR = pr.findall(str(first))
# print PR
PR
```

#### 5) Lastly, get Average Pre Tournament Chess Rating of Opponents

```python
# create i and n
i = 0
n = 0
# create a empty list AR = Average Rating
AR = []
# create a while loop
while(i<len(first)):
    # create a regular expression pattern and store it in t
    t = re.compile('[0-9]+\|')
    # use findall function find all values which match the pattern
    num = t.findall(str(first[i]))
    # create a regular expression pattern and store it in w
    w = re.compile('[0-9]+')
    # use findall function 
    num = w.findall(str(num))
    #create a for loop for gathering total points
    for x in range(len(num)):
        n = n + int(NP[int(num[x])-1])
    # get the AR and use round function to make it simple                 
    AR.append(round(n/len(num),-1))
    i=i+1
    n = 0
#print AR
AR
```

```python
# Player’s Name, Player’s State, Total Number of Points, Player’s Pre-Rating, and Average Rating of Opponents
Whole = {'Player':PN,'State':state,'Total Points':points,'Pre-Rating':PR,'Avg_Rating':AR} 
# Create DataFrame 
Table = pd.DataFrame(Whole) 
Table
```

```python
# convert data as csv file
Table.to_csv('Whole.csv')
```
